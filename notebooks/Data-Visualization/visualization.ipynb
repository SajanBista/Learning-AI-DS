### Hexibin plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (large number of points)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

plt.hexbin(x, y)
plt.colorbar(label='Counts in bin')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hexbin Plot of Point Density')
plt.show()
```

### Violin plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for multiple groups (as a list of arrays)
data = [np.random.normal(0, 1, 100),
        np.random.normal(2, 1.5, 120),
        np.random.normal(-1, 0.8, 90),
        np.random.normal(3, 2, 90)]

plt.violinplot(data)
plt.xticks(np.arange(1, len(data) + 1), ['Group 1', 'Group 2', 'Group 3', 'Group 4'])
plt.ylabel('Value')
plt.title('Violin Plot of Multiple Groups')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Swarm plots

```python
import seaborn as sns

import pandas as pd

data = pd.DataFrame({
    'semester' : ['1 sem','2 sem', '3 sem', '4 sem', '5 sem'],
    'gpa':  [3.32, 3.16, 3.00,2.9, 3.00]
})
data
sns.swarmplot(x ='semester', y = 'gpa', data = data)
plt.title('Using swarm plot')
plt.xlabel('semester')
plt.ylabel('gpa')

plt.show()
```

### Pair plots

```python
data = pd.DataFrame({
    'sem 1': [3.0, 3.3, 3.0, 3.0, 2.7],
    'sem 2': [ 3, 3, 3, 2.7, 3],
    'sem 3': [2.7, 3.0, 3.0, 3.3, 3],
    'sem 4': [ 2.7, 2.7, 2.7, 3.0, 3.0]
})

sns.pairplot(data)

plt.show()
```

# Seaborn

### Histogram

The displot() function will return a histogram by default. The displot() can also create a KDE or rug plot
which are useful ways to look at the data. Seaborn can also combine these plots so you can perform more meaningful 
analysis.

```python
#importing  the necessary libraries
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
# Generating some sample data(a normal distribution)
np.random.seed(42) 
data_basic = np.random.normal(loc=0, scale =1, size = 1000)

plt.figure(figsize=(8,5))
sns.histplot(data = data_basic)
plt.title('Trying histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.grid(axis ='y',  alpha = 0.75)

plt.show
```
